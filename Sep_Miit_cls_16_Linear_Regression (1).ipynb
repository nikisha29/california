```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
cal_data=fetch_california_housing()
cal_data
```

```python
df=pd.DataFrame(cal_data.data,columns=cal_data.feature_names)
df['MedHouseVal']=cal_data.target
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
sns.scatterplot(x='Latitude',y='Longitude',data=df, hue='MedHouseVal', palette='rainbow')
```

```python
l=['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','MedHouseVal']
for i in l:
    sns.displot(df[i], kde=True)
    plt.show()
```

```python
df['MedInc']=df['MedInc'] * 10000

df['MedHouseVal']=df['MedHouseVal'] * 100000
```

```python
df
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
# split the data into input and target variables
X=df.drop('MedHouseVal', axis=1) # input variables
y=df['MedHouseVal'] # target variable
```

```python
# split the input data and target data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# intiate the model
from sklearn.linear_model import LinearRegression
model = LinearRegression() # instantiate the model
```

```python
# train the model
model.fit(X_train, y_train) # fit the model to the training data
```

```python
# make predictions for the test data
y_pred = model.predict(X_test)
```

```python
# evaluate the model

# manual evaluation

eval_df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
eval_df['Absolute_error'] = abs(eval_df['Actual'] - eval_df['Predicted'])
eval_df['Squared_error'] = (eval_df['Actual'] - eval_df['Predicted'])**2
print(eval_df['Absolute_error'].mean())
print(eval_df['Squared_error'].mean())
print
```

```python
SSE=(eval_df['Actual'] - eval_df['Predicted'])**2
SST=(eval_df['Actual'] - eval_df['Actual'].mean())**2
r2= 1- (SSE.sum()/SST.sum())
print(r2)
```

```python
# import the metrics module
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 Score:', r2_score(y_test, y_pred))
```
